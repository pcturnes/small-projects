```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d

# Number of points
n_points = 50

# Generate random x and y coordinates between 0 and 1
x = np.random.rand(n_points)
y = np.random.rand(n_points)
# Combine x and y into a single array of points
points = np.column_stack((x, y))

# Compute the Voronoi diagram
vor = Voronoi(points)

# Plot Voronoi tessellation
fig, ax = plt.subplots()
voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='black', line_width=1.5)

# Beautify the plot
ax.set_aspect('equal')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_title('Voronoi Tessellation')
plt.show()
```
